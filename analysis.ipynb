**Classificação de estrelas, galáxias e quasares.**

Na astronomia, a classificação estelar é a classificação de estrelas com base em suas características espectrais. O esquema de classificação de galáxias, quasares e estrelas é um dos mais fundamentais da astronomia. A primeira catalogação de estrelas e sua distribuição no céu levou ao entendimento de que elas compõem nossa própria galáxia e, após a distinção de que Andrômeda era uma galáxia separada para a nossa, numerosas galáxias começaram a ser pesquisadas à medida que telescópios mais poderosos foram construídos. Este data visa classificar estrelas, galáxias e quasares com base em suas características espectrais.

A classificação estelar é essencial para compreender a diversidade e evolução do universo. Ao analisar as características espectrais das estrelas, os astrônomos podem determinar sua temperatura, composição química e estágio evolutivo. Essa informação não apenas fornece insights sobre a própria estrela, mas também contribui para a compreensão mais ampla da formação e evolução das galáxias que as abrigam. À medida que a tecnologia telescópica avança, novas descobertas são feitas, revelando a riqueza e complexidade do cosmos. A classificação de estrelas, galáxias e quasares com base em suas características espectrais continua sendo um pilar na exploração do vasto e fascinante panorama cósmico.

Os dados consistem em 100.000 observações de espaço realizadas pelo SDSS (Sloan Digital Sky Survey). Cada observação é descrita por 17 colunas de características e uma coluna de classe que a identifica como uma estrela, galáxia ou quasar.

```python
# Mapa de Calor
def correlation(data):
    plt.figure(figsize=(13, 13))
    sns.heatmap(data.corr(), annot=True, fmt='.2f',
                annot_kws={'size': 10}, linewidths=0.01)
    plt.title("Data correlations")
    plt.show()
```

```python
correlation(data)
```

- O valor de maior correlação positiva foi entre spec_obj_ID e MJD, porque ambos estão relacionados ao tempo de coleta de dados. # Funções de Visualização de Dados

Uma **Estimativa de Densidade do Núcleo (KDE)** é uma forma não paramétrica de estimar a função de densidade de probabilidade de uma variável aleatória. Ela oferece uma representação suavizada da distribuição subjacente de um conjunto de pontos de dados. Ao contrário de métodos paramétricos que assumem uma forma específica para a distribuição (por exemplo, distribuição normal), o KDE faz menos suposições sobre a forma dos dados.

Definimos duas funções, **plot e log_plot**, para visualizar as distribuições de colunas especificadas no conjunto de dados. Essas funções são úteis para comparar as distribuições entre diferentes classes e para o conjunto de dados como um todo.

**plot(columns):** Esta função cria uma grade de gráficos de Estimação de Densidade do Núcleo (KDE) para as colunas fornecidas, separando as distribuições com base na característica "classe". Organiza os gráficos em uma grade com 3 colunas, calculando o número necessário de linhas para acomodar todas as colunas fornecidas.

**log_plot(columns):** Semelhante à função plot, esta visualiza as distribuições dos logaritmos das colunas fornecidas. Isso pode ser útil para entender as características logarítmicas dos dados.

```python
le = LabelEncoder()
data["class"] = le.fit_transform(data["class"])
data["class"] = data["class"].astype(int)

# Função para plotar a distribuição KDE (Kernel Density Estimate) de uma coluna em relação às classes
def plot(column):
    for i in range(3):
        sns.kdeplot(data=data[data["class"] == i][column], label = le.inverse_transform([i]), fill = True)
    sns.kdeplot(data=data[column],label = ["All"], fill = True)
    plt.legend();

# Função para plotar a distribuição KDE após a aplicação do logaritmo em uma coluna em relação às classes
def log_plot(column):
    for i in range(3):
        sns.kdeplot(data=np.log(data[data["class"] == i][column]), label = le.inverse_transform([i]), fill = True)
    sns.kdeplot(data=np.log(data[column]),label = ["All"], fill = True)
    plt.legend();
```

```python
log_plot('u')
```
